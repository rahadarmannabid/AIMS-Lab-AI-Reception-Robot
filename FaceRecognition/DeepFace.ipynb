```python
from deepface import DeepFace
import cv2
from deepface.basemodels import VGGFace
import matplotlib.pyplot as plt
import pandas
```

```python
def verify(img1_path,img2_path, model_name="DeepFace"):
    img1=cv2.imread(img1_path)
    img2=cv2.imread(img2_path)
    plt.imshow(img1[:,:,::-1])
    plt.show()
    plt.imshow(img2[:,:,::-1])
    
    result=DeepFace.verify(img1_path,img2_path, model_name=model_name)
    print('result', result)
```

```python
verify("./database/myface1.jpg","./database/myface2.jpg", model_name="VGG-Face")
```
